```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
import warnings
```

```python
x,_ = make_moons(n_samples=500, noise=0.06,random_state=0)
```

```python
#Clustering

db = DBSCAN(eps=0.1, min_samples=5)
labels = db.fit_predict(x)
```

```python
#plots

plt.scatter(x[:,0], x[:,1], c=labels, cmap='plasma')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.datasets import make_circles
```

```python
x,_ = make_circles(n_samples=500, noise=0.05, random_state=0, factor=0.5)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
db = DBSCAN(eps=0.1, min_samples=5)
labels = db.fit_predict(x)
```

```python
plt.scatter(x[:,0], x[:,1], c=labels, cmap='rainbow')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
labels = km.fit_predict(x)
```

```python
plt.scatter(x[:,0], x[:,1], c=labels, cmap='rainbow')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.datasets import make_blobs
```

```python
x,_ = make_blobs(n_samples=500, centers=3, cluster_std=0.5, random_state=0)
```

```python
plt.scatter(x[:,0], x[:,1])
```
